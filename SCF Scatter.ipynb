```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('scf-food-cg/sub-data.csv')
df.head()

x = 'KGINC'
y = 'FOODAWAY'

plt.scatter(x = df[x], y = df[y], s = 1)
plt.title('Amount Spent on Restaurants by Capital Gains')
plt.xlabel('Capital Gains (in 100s of millions)')
plt.ylabel('Amount Spent (in dollars)')

```
