```python
import statsmodels.api as sm #import statastical package
import numpy as np

#generate dummy variables

#use pandas get_dummies function
dummies_1 = pd.get_dummies(df['RACE'])
df['IS-W'] = dummies_1[1]

#alter current variable to get dummy
df['IS-F'] = df['HHSEX'] -1
df.head()
```

```python
X = df[['EDCL', 'IS-F', 'IS-W']] #independent variables
y = df['ASSET']/1000000 #dependent variable, assets in millions of dollars

X = sm.add_constant(X)
model = sm.OLS(y, X)
results = model.fit()
results.summary()
```
