```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager
df = pd.read_csv('../congress-terms.csv')
```

```python
#SET STYLE FOR REST OF NOTEBOOK
matplotlib.style.use('ggplot') #set stylesheet (find options here: https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html)
matplotlib.rcParams['font.size'] = 16
matplotlib.rcParams['figure.figsize'] = (15, 12) #standardise figure size
matplotlib.rcParams['font.family'] =  'Franklin Gothic Medium'
#matplotlib.rcParams['font.weight'] =  'bold' #<-consider adding this line (uncommented) if using a thin font
```

```python
#GET Fonts and write to file fonts.txt in same folder (NOTE: not every font will work)
fonts = font_manager.findSystemFonts(fontpaths = None, fontext = 'ttf')
file = open('fonts.txt', 'w')
file.write(str(fonts))
```

```python
plt.scatter(df['congress'], df['age']) #creat scatter plot
plt.xlabel('Congress Term') #add x axis label
plt.ylabel('Congressmember Age') #add y axis label
matplotlib.rcParams['figure.figsize'] = (12, 8) #change fig size
plt.title('Age by Congress Term', fontsize = 24) #add title and adjust font size
plt.ylim([40, 90]) #limit range of y axis
plt.savefig('Congress_Age+Term_Scatter.png', dpi = 300) #save to high res png
```

```python
#create bar chart
cf = df[df['chamber'] == 'house']
table = cf['party'].value_counts(normalize = True).rename_axis('Party').reset_index(name = 'Percent')
table['Percent'] = table['Percent'] * 100
table = table[table['Percent']>30]
plt.barh(table["Party"],table["Percent"],0.7, color = ['#00FFFF', '#009966'])
matplotlib.rcParams['figure.figsize'] = (12, 9)
plt.ylabel('Party', fontsize = 16) 
plt.xlabel('Percentage', fontsize = 16) #add x label and adjust font size
plt.title('Party Representation in House (going way back)', fontsize = 20) #add title and adjust font size
plt.savefig('Congress_Party_Bar.png', dpi = 300) #save to high res png
```

```python
#create pie chart
matplotlib.rcParams['figure.figsize'] = (9, 9)
plt.pie(table['Percent'], labels = table['Party'], textprops={'fontsize': 20}, autopct='%1.1f%%') #create pie chart w labels and percentage labels
plt.title('Party Representation in House (going way back)')
plt.savefig('Congress_Party_Pie.png', dpi = 300) #save to high res png
```
