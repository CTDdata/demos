```python
import pandas as pd
file = 'SCDB_2021_01_caseCentered_Docket.dta'
df = pd.read_stata(file, convert_categoricals = False)
df.head()
```

```python
#assign year and month using datetime (dt)
date_col = 'dateDecision'
df['year'] = df[date_col].dt.year
df['month'] = df[date_col].dt.month
df[[date_col,'year','month']].head()

```

```python
#assign year using string slice
date_col = 'caseId'
df['year'] = df[date_col].str[:4]
df[[date_col,'year']].head()
```

```python
#df['year'] = pd.to_numeric(df['year']) #<-convert to numeric IF string datatype
after_50_df = df[df['year'] >= 1950]
after_50_before_70_df = after_50_df[after_50_df['year']<1970]
min_year = after_50_before_70_df['year'].min()
max_year = after_50_before_70_df['year'].max()
print(min_year, max_year)
```
